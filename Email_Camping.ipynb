```python
# Here I Assign The Value Like "Morning":0,"Noon":1,"Evening":2

df_email_camping.times_of_day = df_email_camping.times_of_day.replace({"Morning":0,"Noon":1,"Evening":2}) 
```

```python
# In Is_Price Column Here 99% Of Values Zero Thats The Reason We Remove This Column
df_email_camping.drop(['is_price'],axis=1,inplace=True)
```

```python
# In campaign_id There Is All Values Unique So Also We Remove The Column
df_email_camping.drop(['campaign_id'],axis=1,inplace=True)
```

```python
df_email_camping
```

```python
# Outliers Tritment
sns.boxplot(df_email_camping.body_len)
```

```python
# Copy Data Frame
df_email_campaign_outliers = df_email_camping.copy()
```

```python
# Process For Overcom Outliers
Q1 = np.percentile(df_email_campaign_outliers["body_len"],25)
Q3 = np.percentile(df_email_campaign_outliers["body_len"],75)
IQR = Q3-Q1
lower_fence = Q1-(1.5*IQR)
upper_fence = Q3+(1.5*IQR)
print(lower_fence,upper_fence)
```

```python
# Above We Found lower_fence and upper_fence for outliers finding
```

```python
null_list = []
for i in df_email_campaign_outliers["body_len"]:
    if i <= lower_fence or i >= upper_fence:
        null_list.append(i)
print(null_list)
```

```python
# Above We Diplay The Outliers Using Applying Lower_fance And Upper_Fance
```

```python
for i in df_email_campaign_outliers["body_len"]:
    if i <= lower_fence or i >= upper_fence:
        df_email_campaign_outliers. drop(df_email_campaign_outliers.index[df_email_campaign_outliers['body_len'] == i], inplace=True)
```

```python
# Drop The Outliers From The 
df_email_campaign_outliers
```

```python
mean_for_body_len=df_email_campaign_outliers["body_len"].mean()
print(mean_for_body_len)
```

```python
for i in df_email_camping["body_len"]:
    if i <= lower_fence or i >= upper_fence:
        df_email_camping.body_len.replace(i,mean_for_body_len,inplace=True)
```

```python
sns.boxplot(df_email_camping["body_len"])
```

```python
null_list = []
for i in df_email_camping["body_len"]:
    if i <= lower_fence or i >= upper_fence:
        null_list.append(i)
print(null_list)
```
